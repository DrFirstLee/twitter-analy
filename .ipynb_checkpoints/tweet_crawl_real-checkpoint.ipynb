````python
import requests
import os
import json
import mykeys
import pandas as pd
import datetime
import time
def bearer_oauth(r):
    """
    Bearer Token ìœ¼ë¡œ oauth ê¶Œí•œ ë°›ìŒ!! rì„ ë¦¬í„´í•´ì„œ ì´í›„ headerì— ë„£ëŠ”ë‹¤!!
    """
    r.headers["Authorization"] = f"Bearer {bearer_token}"
    r.headers["User-Agent"] = "v2UserLookupPython"
    return r

bearer_token = mykeys.Bearer_Token

import openai
import json

openai.api_key =mykeys.openai_key  # API Key

## ì—¬ê¸°ì— ì‚¬ìš©ì ì•„ì´ë””ì…ë ¥!! : elonmush
usernames = "usernames=nvidia,NVIDIAAI,elonmusk,CoinMarketCap,RayDalio,AndrewYNg,realDonaldTrump,SpaceX,JeffBezos,OpenAI,BillGates,VitalikButerin,ARKInvest,CathieDWood,BillAckman,sama,binance,RaoulGMI,ylecun,brian_armstrong,researchhub,Tesla,APompliano,PeterSchiff"
user_fields = "user.fields=description,created_at"

user_info_url = "https://api.twitter.com/2/users/by?{}&{}".format(usernames, user_fields)

response = requests.request("GET", user_info_url, auth=bearer_oauth,)
print(response.status_code)

user_json_response = response.json()
user_json_response
````

````python
user_id_l = []
user_name_l = []
for users_info in user_json_response['data']:
    user_id_l.append( users_info['id'])
    user_name_l.append(users_info['username'] )
    print(users_info['username'] + " : " + users_info['id'])
df_user = pd.DataFrame()
df_user['user_id'] = user_id_l
df_user['user_name'] = user_name_l
df_user
````

````python
df_final = pd.DataFrame()
for idx, sr in df_user.iterrows():
    
    user_id =df_user.loc[idx,'user_id']
    user_tweets_url = "https://api.twitter.com/2/users/{}/tweets".format(user_id)
    # ìµœê·¼ 24ì‹œê°„ì˜ íŠ¸ìœ—ì„ ê°€ì ¸ì˜¤ê¸° ìœ„í•œ start_time ì„¤ì • (RFC3339 í˜•ì‹)
    now = datetime.datetime.utcnow()  # í˜„ì¬ ì‹œê°„ (UTC)
    start_time = (now - datetime.timedelta(days=1)).strftime("%Y-%m-%dT%H:%M:%SZ")  # RFC3339 í˜•ì‹ìœ¼ë¡œ ë³€í™˜

    # 
    # ìš”ì²­ íŒŒë¼ë¯¸í„° ì„¤ì • (íŠ¸ìœ— ì‘ì„± ì‹œê°„ í•„ë“œ ë° ì‹œì‘ ì‹œê°„ ì„¤ì •)
    target_params = {
       "tweet.fields": "created_at,public_metrics",
        "start_time": start_time,
        "max_results": 100  # ìµœëŒ€ 100ê°œì˜ íŠ¸ìœ—ì„ ìš”ì²­
    }
    response = requests.request("GET", user_tweets_url, auth=bearer_oauth, params=target_params)

    json_response =  response.json()
    json_response
    if 'data' in json_response:
        df = pd.DataFrame(json_response['data'])
        # 'created_at'ì„ datetime í˜•ì‹ìœ¼ë¡œ ë³€í™˜
        df['created_at2'] = pd.to_datetime(df['created_at'])
        # 'created_at'ì„ ì„œìš¸ ì‹œê°„(UTC+9)ìœ¼ë¡œ ë³€í™˜í•˜ê³ , ì›í•˜ëŠ” í˜•ì‹ìœ¼ë¡œ ì¶œë ¥
        df['created_at2'] = df['created_at2'].dt.tz_convert('Asia/Seoul').dt.strftime('%Y%m%d-%H%M%S')
        df['user_id'] = sr['user_id']
        df['user_name'] = sr['user_name']
        df_final= pd.concat([df_final, df])
        print(idx, df_user.loc[idx,'user_id'] , df_user.loc[idx,'user_name'], len(df))
        time.sleep(1)
    else:
        print(idx, df_user.loc[idx,'user_id'] , df_user.loc[idx,'user_name'], 'skipped')
````
